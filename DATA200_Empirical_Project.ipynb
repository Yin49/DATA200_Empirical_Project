```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data

bmi = pd.read_sas('BMX_H.xpt', format = 'xport') 
diet1 = pd.read_sas('DR1IFF_H.xpt', format = 'xport')
diet2 = pd.read_sas('DR2IFF_H.xpt', format = 'xport')

display(bmi.head())
display(diet1.head())
display(diet2.head())
```

```python
col_names = ['SEQN', 'energy', 'protein', 'carbonhydrate', 'sugar', 'fiber', 'fat', 'sat_fatty_acid', 
             'mono_fatty_acid', 'poly_fatty_acid', 'cholesterol', 'caffeine', 'alcohol', 'moisture', 'infant']
```

```python
## Clean dataset -- diet1

diet1 = diet1[['SEQN','DR1IKCAL', 'DR1IPROT', 'DR1ICARB', 'DR1ISUGR', 'DR1IFIBE', 
              'DR1ITFAT', 'DR1ISFAT', 'DR1IMFAT', 'DR1IPFAT', 'DR1ICHOL',
              'DR1ICAFF', 'DR1IALCO', 'DR1IMOIS', 'DRABF']]
diet1.columns = col_names
```

```python
# exclude infants
diet1 = diet1[diet1['infant'] == 2]
```

```python
diet1 = diet1.drop(['infant'], axis = 1)
```

```python
# Group by
diet1 = diet1.groupby(by = ['SEQN']).sum()
```

```python
## Clean dataset -- diet2

diet2 = diet2[['SEQN','DR2IKCAL', 'DR2IPROT', 'DR2ICARB', 'DR2ISUGR', 'DR2IFIBE', 
              'DR2ITFAT', 'DR2ISFAT', 'DR2IMFAT', 'DR2IPFAT', 'DR2ICHOL',
              'DR2ICAFF', 'DR2IALCO', 'DR2IMOIS', 'DRABF']]
diet2.columns = col_names

# exclude infants
diet2 = diet2[diet2['infant'] == 2]

diet2 = diet2.drop(['infant'], axis = 1)

# Group by
diet2 = diet2.groupby(by = ['SEQN']).sum()
```

```python
## Clean dataset -- bmi
bmi = bmi[['SEQN', 'BMXBMI']]
```

```python
bmi = bmi.dropna()
```

```python
bmi.columns = ['SEQN', 'bmi']
```

```python
bmi['status'] = bmi['bmi']

for i in bmi.index:
    if bmi.loc[i, 'bmi'] < 18.5:
        bmi.loc[i, 'status'] = 0
    elif bmi.loc[i, 'bmi'] < 25.0:
        bmi.loc[i, 'status'] = 1
    elif bmi.loc[i, 'bmi'] < 30.0:
        bmi.loc[i, 'status'] = 2
    elif bmi.loc[i, 'bmi'] < 35.0:
        bmi.loc[i, 'status'] = 3
    elif bmi.loc[i, 'bmi'] < 40.0:
        bmi.loc[i, 'status'] = 4
    else:
        bmi.loc[i, 'status'] = 5
```

```python
# reindex
bmi = bmi.set_index('SEQN')

bmi.index
```

```python
# merge diet1 & diet2
diet = pd.concat([diet1,diet2],axis=1)
```

```python
diet = diet.dropna()
```

```python
diet = diet.groupby(level=0,axis=1).mean()
```

```python
# merge -- diet & bmi
diet_bmi = pd.concat([diet,bmi],axis=1)
diet_bmi = diet_bmi.dropna()
```

```python
def loading_plot(coeff, labels):
    n = coeff.shape[0]
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1], head_width = 0.05, head_length = 0.05, color = '#21918C',alpha = 0.5)
        plt.text(coeff[i,0]* 1.5, coeff[i,1] * 1.5, labels[i], color = '#21918C', ha = 'center', va = 'center')
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.grid()
```

```python
fig, ax = plt.subplots(figsize = (15,15))
loading_plot(pca.components_.T, col_names_X)
```
